## Start Cutting 20 hits per bin

```python
counts, edges, plot = pp.hist(time[np.where(np.logical_and(np.logical_and(channel==2,edge==0),time<=endTime))],bins=60*1000,label="Firsy", histtype="barstacked")

lowOccuranceBins = []
otherBins = []
cutOff = 20
pp.axhline(cutOff, color = 'gold')
pp.show()
for _ in range(len(counts)):
    occ = counts[_]
    if (occ > 0 and occ <= cutOff):
        lowOccuranceBins.append([_, occ])
    else:
        otherBins.append([_, occ])

    
print(len(counts))
print()
print(len(lowOccuranceBins))
print()
print(len(otherBins))
```

```python
lowTimeRanges = []
for pair in lowOccuranceBins:
    time1 = pair[0]
    #print(time1)
    lowTimeRanges.append([(time1)*1000000,(time1+1)*1000000])
print("~~~~~")
print(len(lowTimeRanges))
```

```python
def lowTimeBool(time, lowTimeRanges): #Returns true or false if that time is in the range
    for rang in lowTimeRanges:
        if( np.logical_and(time>=rang[0],time<rang[1])):
            display("true")
            return True
    display("false")
    return False
```

```python
def lowTimeFile(time, lowTimeRanges, tmax = 60000000000.0): #Sorts times in/out of rangeand returns 2 arrays
    start = datetime.now()
    lowTimeArray = []
    otherTimeArray = []
    for _ in range(len(time)):
        val = False
        clear_output(wait=True)
        display(str(round(time[_]*100 / tmax, 3)) + " %")
        for rang in lowTimeRanges:
            if( np.logical_and( time[_] >= rang[0], time[_] < rang[1] ) ):
                val = True
        if(val):
            lowTimeArray.append(time[_])
        else:
            otherTimeArray.append(time[_])
    #lowTimeArray = np.array(lowTimeArray)
    #otherTimeArray = np.array(otherTimeArray)
    end = datetime.now()
    print("DONE (-).* " + str(end-start))
    return lowTimeArray, otherTimeArray
    
```

```python
tmax = 60000000000.0
```

```python
lowTimes, highTimes = lowTimeFile(time[np.logical_and(np.logical_and(channel==cn1,edge==0), time<=tmax)], lowTimeRanges)
```

```python
print(len(lowTimes))
print()
print(len(highTimes))
```

## Time Diff Between Consecutive

```python
#threhshold = 0.5
bns = 200
one_s_bins, *trash = np.histogram(lowTimes,bins=6000)
#two_s_bins, *trash = np.histogram(highTimes,bins=60)

pp.hist(lowTimes, bins = bns, label = "Low Times", histtype = "step")
pp.hist(highTimes, bins = bns, label = "Low Times", histtype = "step")
#pp.axvline(x=threhshold,ymin=0,ymax=600, color = "gold")
pp.legend()
pp.xlabel('x')
pp.ylabel("Count")
pp.title("Dont know (" + str(bns)+" bins)")
pp.show()

pp.hist(one_s_bins, bins = bns, label = "Low Times", histtype = "step")
#pp.axvline(x=threhshold,ymin=0,ymax=600, color = "gold")
pp.legend()
pp.xlabel('x')
pp.ylabel("Count")
pp.title("Dont know (" + str(bns)+" bins)")
pp.show()
```

```python
def getDiffs(tymes):
    diffs = []
    for _ in range(len(tymes) - 1):
        diffs.append(tymes[_+1] - tymes[_])
    return diffs
```

```python
lowDiffs = getDiffs(lowTimes)
highDiffs = getDiffs(highTimes)
allDiffs = lowDiffs + highDiffs
```

```python
display(len(lowDiffs))
display()
display(len(highTimes))
```

```python
bns = 100
ax = pp.subplot(1,1,1)
ax.hist(np.log10(highDiffs), bins = bns, histtype = 'step', density = True, label = 'High rate')
ax.hist(np.log10(lowDiffs), bins = bns, histtype = 'step', density = True, label = 'Low rate')
ax.hist(np.log10(allDiffs), bins = bns, histtype = 'step', density = True, label = 'All')
ax.grid()
ax.set_axisbelow(True)
#ax.set_xlim(1,7)
ax.set_ylim(-0.05,0.8)
ax.legend(loc = 'upper left')
ax.set_ylabel("Normalized Counts")
ax.set_xlabel(r"$\mathrm{log}_{10} \left(\Delta t/\mathrm{ns}\right)$")
ax.set_title("Normalized Time Difference between Consecutive Hits\n 1-20 hits per ms")
pp.show()
```
